```python
llm=ChatGroq(api_key=GROQ_API_KEY,model_name='deepseek-r1-distill-qwen-32b',temperature=0.7)
```

```python
llm.invoke('which is the best place to travel in india')
```

```python
from langchain.prompts import PromptTemplate  

template = """  
You are an AI travel planner. Create a personalized **day-by-day travel itinerary** based on the user's preferences.  
Consider real-time factors like **weather, local events, and budget constraints**.  

### **User Details:**  
- **Destination:** {destination}  
- **Travel Dates:** {dates}  
- **Budget:** {budget}  
- **Preferences:** {preferences} (e.g., adventure, culture, relaxation, food, nightlife)  
- **Number of Travelers:** {num_travelers}  
- **Accommodation Type:** {accommodation} (hotel, Airbnb, hostel)  

### **Your Task:**  
- Create a **detailed daily plan** for the user.  
- Include **morning, afternoon, and evening activities**.  
- Suggest **places to visit, restaurants, transport options, and estimated costs**.  
- Consider **weather conditions and local events**.  
- Ensure the itinerary **matches the budget and preferences**.  

### **Format:**  
**Day 1:**  
ðŸŒ… Morning: [Activity] at [Location] â€“ [Brief Description]  
ðŸŒž Afternoon: [Activity] at [Location] â€“ [Brief Description]  
ðŸŒ™ Evening: [Activity] at [Location] â€“ [Brief Description]  

**Day 2:** ...  

Return a structured JSON format for API use.  

"""

prompt = PromptTemplate(
    input_variables=["destination", "dates", "budget", "preferences", "num_travelers", "accommodation"],
    template=template
)

# Example Usage
filled_prompt = prompt.format(
    destination="Paris",
    dates="March 15 - March 20",
    budget="Medium",
    preferences="Culture, Food, Landmarks",
    num_travelers="2",
    accommodation="Hotel"
)

print(filled_prompt)
```

```python
from langchain.chains import LLMChain
chain = LLMChain(llm=llm, prompt=prompt)

# Run the Chain with Example Data
response = chain.run({
    "destination": "Paris",
    "dates": "March 15 - March 20",
    "budget": "Medium",
    "preferences": "Culture, Food, Landmarks",
    "num_travelers": "2",
    "accommodation": "Hotel"
})

print(response)
```

```python
chain = prompt | llm  

# Define user input
user_input = {
    "destination": "Paris",
    "dates": "March 15 - March 20",
    "budget": "Medium",
    "preferences": "Culture, Food, Landmarks",
    "num_travelers": "2",
    "accommodation": "Hotel"
}

# Run the chain
response = chain.invoke(user_input)

print(response)
```
